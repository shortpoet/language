```python
GI_dict_list = []
for x in range(1, 270):
    page = x
    url = f'http://www.glycemicindex.com/foodSearch.php?ak=list&food_name_search_type=cn&food_name=&gi_search_type=lte&gi=&gl_search_type=lte&gl=&lop=OR&find=Find+Records&page={page}'
    driver.get(url)
    html = driver.page_source
    gi_soup1 = bs(html, 'html.parser')
    cells = gi_soup1.find_all('p', class_='body')
    cells = cells[2:]
    foods = [cell.text for cell in cells[::5]]
    GI = [cell.text for cell in cells[1::5]]
    serve_g = [cell.text for cell in cells[2::5]]
    carb_per_serve_g = [cell.text for cell in cells[3::5]]
    GL = [cell.text for cell in cells[4::5]]
    print(url)
    for x in range(10):
        GI_dict_list.append({
            'food': foods[x], 'GI': GI[x], 'serve_g': serve_g[x], 'carb_per_serve_g': carb_per_serve_g[x], 'GL': GL[x]
        })
        
```
